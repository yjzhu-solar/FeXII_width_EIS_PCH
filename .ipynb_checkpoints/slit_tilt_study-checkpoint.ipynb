```python
#A Gaussian fitting model
class myGaussian1D(Fittable1DModel):
    amplitude = Parameter()
    mean = Parameter()
    stddev = Parameter()
    bg = Parameter()

    @staticmethod
    def evaluate(x, amplitude, mean, stddev,bg):
        return amplitude * np.exp((-(1 / (2. * stddev**2)) * (x - mean)**2)) + bg

    @staticmethod
    def fit_deriv(x, amplitude, mean, stddev,bg):
        d_amplitude = np.exp((-(1 / (stddev**2)) * (x - mean)**2))
        d_mean = (2 * amplitude *
                  np.exp((-(1 / (stddev**2)) * (x - mean)**2)) *
                  (x - mean) / (stddev**2))
        d_stddev = (2 * amplitude *
                    np.exp((-(1 / (stddev**2)) * (x - mean)**2)) *
                    ((x - mean)**2) / (stddev**3))
        return [d_amplitude, d_mean, d_stddev,np.ones_like(d_amplitude)]
```

```python
class myQuadra(Fittable1DModel):
    c1 = Parameter()
    c2 = Parameter()
    
    @staticmethod
    def evaluate(x, c1, c2):
        return c2 * np.square(x) + c1*x - 512*c1 - np.square(512)*c2
    
    @staticmethod
    def fit_deriv(x, c1, c2):
        d_c1 = x - 512
        d_c2 = np.square(x) - np.square(512)
        return [d_c1, d_c2]
```

```python
class FeXII_fit:
    def __init__(self,data,wvl,p0):
        self.data = data
        self.wvl = wvl
        self.shape = data.shape
        self.result = np.tile(p0,(self.shape[0],1))
                
    def run_lsq(self):
        for ii in range(self.shape[0]):
            g_init = myGaussian1D(amplitude=self.result[ii,1], mean=self.result[ii,0], stddev = self.result[ii,2],bg = 0.)
            fit_g = fitting.LevMarLSQFitter()
            g = fit_g(g_init, self.wvl, self.data[ii,:])
            self.result[ii,:] = np.array([g.mean.value, g.amplitude.value, np.abs(g.stddev.value), g.bg.value])

    def run_mcmc(self):
        pass
    
    def para(self):
        return np.array([np.array([p_[0], np.sqrt(2)*p_[1]*p_[2], 2.355*p_[2]]) for p_ in self.result])
    
    def show(self):
        for ii, p_ in enumerate(self.result):
            display(Math("\lambda = {:.3f}\  \mathrm{{Int}} = {:.2e}\  \mathrm{{FHWM}} = {:.1e}".format(p_[0],
                                                                                                        np.sqrt(2)*p_[1]*p_[2],
                                                                                                        2.355*p_[2])))
    def fit_para(self):
        centroid = np.array([p_[0] for p_ in self.result])
        centroid_rel = centroid - centroid[0]
        ypix = np.arange(512,1024)
        
        #quadra_init = myQuadra(c1=0.14e-3, c2=0.02e-6)
        quadra_init = models.Polynomial1D(degree=2,c0 = 0.,c1=0.14e-3, c2=0.02e-6)
        quadra.c0.fixed = True
        
        fit_quadra = fitting.LevMarLSQFitter()
        quadra = fit_quadra(quadra_init,ypix,centroid_rel)
        print(quadra)
        
        fig, ax = plt.subplots(figsize=(8,6))
        ax.scatter(ypix,centroid_rel)
        ln, = ax.plot(ypix,quadra(ypix),color = '#F596AA',linewidth=2,label = '')
        ax.set_xlabel(r'',fontsize=24)
        ax.set_ylabel(r'',fontsize=24)
        ax.tick_params(labelsize=18)
        plt.subplots_adjust(bottom = 0.1,left=0.1,top=0.97,right=0.97)
```
