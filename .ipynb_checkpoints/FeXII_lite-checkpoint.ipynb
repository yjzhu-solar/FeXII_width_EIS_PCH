```python
#correct slit tilt
def correct_tilt(wvl_old,data_input):
    data_2d = data_input.copy()
    #c1 = 0.0002037
    #c2 = -4.6065e-8
    c1 = 0.1415e-3
    c2 = -0.0205e-6
    
    for ii in range(52):
        #wvl_cor = c1*(ii+512) + c2*((ii+512)**2) - c1*512. - c2*512.**2
        wvl_cor = c1*ii*10 + c2*np.square(ii*10)
        data_2d[ii,:] = np.interp(wvl_old, wvl_old - wvl_cor,data_2d[ii,:])
        
    return data_2d
```

```python
data_limb = data_set['final_limb']
data_disk = data_set['final_disk']
data_limb_cor = correct_tilt(data_set["wvl"],data_limb[:52,:])
data_disk_cor = correct_tilt(data_set["wvl"],data_disk[:52,:])

data_stray_count= np.sum(np.where(data_disk_cor[15:,]>-1.e6,1,0),axis=0)
data_stray = 0.02 * np.sum(data_disk_cor[15:,],axis=0)/data_stray_count
data_stray = np.nan_to_num(data_stray,neginf=0.,posinf=0.)

data_limb_192 = data_limb_cor[:,1160:1190][:,5:23]
data_limb_193 = data_limb_cor[:,1210:1245][:,7:24]
data_limb_195 = data_limb_cor[:,1285:1315][:,5:20]

wvl_192 = data_set["wvl"][1160:1190][5:23]
wvl_193 = data_set["wvl"][1210:1245][7:24]
wvl_195 = data_set["wvl"][1285:1315][5:20]


```
